## 9. Identificação e Remoção de Variáveis Irrelevantes

Foram identificadas variáveis sem valor preditivo:

- `EmployeeNumber`: identificador único do colaborador.
- `EmployeeCount`: valor constante (sempre 1).
- `Over18`: valor constante (sempre Y).
- `StandardHours`: valor constante (sempre 80).

Estas variáveis não contribuem para a capacidade preditiva do modelo e foram removidas para evitar ruído desnecessário.

*Autor: Figueira, L.*

```python
# Identificar colunas com variância zero (valor único)
print("Colunas constantes:", df.columns[df.nunique() == 1].tolist())

# Remover variáveis irrelevantes
colunas_remover = ["EmployeeNumber", "EmployeeCount", "Over18", "StandardHours"]
df = df.drop(columns=colunas_remover)

print(f"\nColunas removidas com sucesso. O dataset passou a ter {df.shape[1]} variáveis.")
# Autor: Figueira, L.
```

## 10. Separação de Variáveis Numéricas e Categóricas

```python
# Separação de Variáveis Numéricas e Categóricas 
variaveis_numericas = df.select_dtypes(include=["int64", "float64"]).columns
variaveis_categoricas = df.select_dtypes(include=["string"]).columns

print(f"Variáveis numéricas: {len(variaveis_numericas)}")
print(f"Variáveis categóricas: {len(variaveis_categoricas)}")
# Autor: Figueira, L.
```

## 11. Estatística Descritiva

```python
# Estatística Descritiva
df.describe()
# Autores: Figueira, L., Afonso, M. e Ferreira, M.
```

## Observações - Estatística Descritiva

A análise das estatísticas descritivas permite identificar dois aspetos relevantes para a fase de modelação:

**Escalas heterogéneas entre variáveis:** `MonthlyIncome` opera numa escala de ~1 000 a ~20 000, enquanto variáveis ordinais como `JobSatisfaction` ou `WorkLifeBalance` variam apenas entre 1 e 4. Esta heterogeneidade torna necessária a aplicação de normalização (e.g., `StandardScaler`) antes do treino de modelos sensíveis à escala, como Regressão Logística, SVM ou KNN.

**Assimetria em variáveis de antiguidade:** `YearsAtCompany`, `YearsInCurrentRole` e `TotalWorkingYears` apresentam mediana consideravelmente inferior à média, indicando distribuições assimétricas à direita - a maioria dos colaboradores tem poucos anos de empresa e uma minoria tem carreiras muito longas.

## 12. Valores Mínimos e Máximos das Variáveis Numéricas

```python
# Amplitude das Variáveis Numéricas (Min/Max)
min_max_df = df[variaveis_numericas].agg(["min", "max"]).transpose()
print("--- Amplitude das Variáveis Numéricas (Min/Max) ---")
print(min_max_df)
# Autor: Afonso, M.
```

## 13. Distribuição da Variável Alvo

```python
# Distribuição da Variável Alvo 
attrition_perc = df["Attrition"].value_counts(normalize=True) * 100
print("Distribuição da Variável Alvo (Desequilíbrio de Classes)")
print(f"  Fica (No):  {attrition_perc['No']:.2f}%")
print(f"  Sai (Yes): {attrition_perc['Yes']:.2f}%")
# Autor: Afonso, M.
```

## Observações - Desequilíbrio de Classes

A variável alvo `Attrition` apresenta uma proporção de aproximadamente **16% (Yes)** face a **84% (No)**, caracterizando um dataset com desequilíbrio de classes significativo.

Este resultado tem implicações diretas nas decisões metodológicas das fases seguintes:

- **Métrica de avaliação:** a *Accuracy* torna-se uma métrica enganosa - um modelo que preveja sempre "No" atingiria 84% de acurácia sem qualquer capacidade preditiva real.
